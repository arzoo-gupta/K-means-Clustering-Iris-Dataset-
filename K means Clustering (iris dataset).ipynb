## Task2 - Predict the optimum number of clusters and represent it visually

```python
#importing library and dataset
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
iris.feature_names
```

```python
df = pd.DataFrame(iris.data , columns = iris.feature_names)
df
```

```python
x=iris.data
x
```

# Define target 

```python
iris.target_names
df['target']=iris.target

##Create a new column inplace of existing column .apply is used 
df['Flower_names']= df.target.apply(lambda x : iris.target_names[x])
df
```

# Let's represent our data through scatter plot

```python
## Creating 3 different dataframe for different 3 flower
df0 = df[:50]
df1 = df[50:100]
df2 = df[100:150]

import matplotlib.pyplot as plt
%matplotlib inline

plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.scatter(df0['sepal length (cm)'] , df0['sepal width (cm)'] , color = 'green', marker = '+')
plt.scatter(df1['sepal length (cm)'] , df1['sepal width (cm)'] , color = 'blue', marker = '.')
```

```python
import seaborn as sns
sns.pairplot(df.drop(['target'],axis=1), hue='Flower_names' , height=2.5, markers = ["8","s","D"])
```

# Now, estimating the value of k

```python
from sklearn.cluster import KMeans

km = KMeans(n_clusters = 5 , init = 'k-means++' , random_state=0)
y = km.fit_predict(x)
print(y)
```

```python
km.cluster_centers_
```

```python
plt.scatter(x[y == 0,0], x[y==0,1], s=15, c= 'red' , label = 'Cluster_I')
plt.scatter(x[y == 1,0], x[y==1,1], s=15, c= 'green' , label = 'Cluster_II')
plt.scatter(x[y == 2,1], x[y==2,1], s=15, c= 'yellow' , label = 'Cluster_III')
plt.scatter(x[y == 3,1], x[y==3,1], s=15, c= 'blue' , label = 'Cluster_IV')
plt.scatter(x[y == 4,1], x[y==4,1], s=15, c= 'cyan' , label = 'Cluster_V')

plt.scatter(km.cluster_centers_[:,0] , km.cluster_centers_[:,1], s = 25 , c = 'magenta' , label = 'Centroids')
plt.legend()
plt.show()
```

## What if there are more features then we will use elbow technique


```python
Error=[]
k_rng = range(1,11)
for k in k_rng:
    km = KMeans(n_clusters = k , init = 'k-means++' , max_iter=300 , n_init=10 , random_state = 0).fit(x)
    km.fit(x)  
    Error.append(km.inertia_)
    
plt.plot(range(1,11) , Error)
plt.title("Elbow Method with range 1-11")
plt.xlabel('Number of Clusters')
plt.ylabel('Error')
plt.show()
```

## Now , Let's fit and predict by taking value 3 

```python
km_1 = KMeans(n_clusters = 3 , random_state = 21)
y = km_1.fit_predict(x)
y
```

```python
km_1.cluster_centers_
```

```python
plt.scatter(x[y == 0,0], x[y==0,1], s=15, c= 'red' , label = 'Cluster_I')
plt.scatter(x[y == 1,0], x[y==1,1], s=15, c= 'green' , label = 'Cluster_II')
plt.scatter(x[y == 2,1], x[y==2,1], s=15, c= 'yellow' , label = 'Cluster_III')

plt.scatter(km.cluster_centers_[:,0] , km.cluster_centers_[:,1], s = 25 , c = 'magenta' , label = 'Centroids')
plt.legend()
plt.show()
```
